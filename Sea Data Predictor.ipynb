```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

def draw_plot():
    # Read data from file
    df = pd.read_csv('epa-sea-level.csv')

    # Making Dataframes for regression lines
    df_year_reg = pd.DataFrame(np.arange(1880, 2051), columns= ['Year'])
    df_year_reg2 = pd.DataFrame(np.arange(2000,2051), columns= ['Year'])
    df_recent = df.copy()
    df_recent = df_recent.iloc[120: 134]


    # Create scatter plot
    fig, ax= plt.subplots(figsize= (12,6))
    ax = plt.scatter('Year', 'CSIRO Adjusted Sea Level', data= df)
    plt.xlim(1870,2060)
    

    # Create first line of best fit
    reg1 = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
    plt.plot( df_year_reg['Year'], reg1.intercept + reg1.slope*df_year_reg['Year'],'r')


    # Create second line of best fit
    reg2 = linregress(df_recent['Year'], df_recent['CSIRO Adjusted Sea Level'])
    plt.plot( df_year_reg2['Year'], reg2.intercept + reg2.slope*df_year_reg2['Year'],'b')

    # Add labels and title
    plt.xlabel('Year')
    plt.ylabel('Sea Level (inches)')
    plt.title('Rise in Sea Level')
    
    # Save plot and return data for testing (DO NOT MODIFY)
    plt.savefig('sea_level_plot.png')
    return plt.gca()
```

```python
draw_plot()
```
